```python
# id:19--38-19
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("week2.csv")
print(df.head())
X1=df.iloc[:,0]
X2=df.iloc[:,1]
X=np.column_stack((X1,X2))
y=df.iloc[:,2]
```

```python
colors = {'1':'blue', '-1': 'green'}
marker = {'1': '+', '-1': 'o'}

plt.figure(figsize=(10,10))
for i in range(len(X1)):
    plt.scatter(X1[i], X2[i], color=colors[str(y[i])], marker=marker[str(y[i])])
plt.legend(['1', '-1'], loc='upper right')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.title('Actual Values')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X, y)
Y = model.predict(X)
b = model.intercept_[0]
w1, w2 = model.coef_.T
print(b)
model.coef_
```

```python
colors1 = {'1':'red', '-1': 'purple'}
marker1 = {'1': 'x', '-1': 's'}

c = -b/w2
m = -w1/w2

plt.figure(figsize=(10,10))
for i in range(len(X1)):
    plt.scatter(X1[i], X2[i], color=colors1[str(y[i])], marker=marker1[str(y[i])])
    
xmin, xmax = -1, 1
ymin, ymax = -1, 1
xd = np.array([xmin, xmax])
yd = m*xd + c
line = plt.plot(xd, yd, label='Decision Boundary')
blue = plt.fill_between(xd, yd, ymin, color='tab:blue', alpha=0.2)
orange = plt.fill_between(xd, yd, ymax, color='tab:orange', alpha=0.2)

markersA = plt.Line2D([0,0],[0,0],color=colors1['1'], marker=marker1['1'], linestyle='')
markersB = plt.Line2D([0,0],[0,0],color=colors1['-1'], marker=marker1['-1'], linestyle='')
plt.legend([markersA, markersB, line[0], blue, orange], ["Actual 1", "Actual -1", "Decision Boundary", "Predicted 1", "Pridected -1"], numpoints=1, loc='upper right')

plt.xlabel('x_1')
plt.ylabel('x_2')
plt.title('Logistic Regression - Predicted Values')
plt.show()
```

```python
from matplotlib.colors import ListedColormap

colors1 = {'1':'red', '-1': 'purple'}
marker1 = {'1': 'x', '-1': 's'}

plt.figure(figsize=(10,10))

for i in range(len(X1)):
    plt.scatter(X1[i], X2[i], color=colors1[str(y[i])], marker=marker1[str(y[i])])

X_1, X_2 = np.meshgrid(np.arange(start = X1.min(), stop = X1.max(), step = 0.01), np.arange(start = X2.min(), stop = X2.max(), step = 0.01))

X_Sq1 = np.column_stack((X_1.ravel(), X_2.ravel(), np.array(np.square(X_1)).ravel(), np.array(np.square(X_2)).ravel()))
plt.contourf(X_1, X_2, model.predict(X_Sq1).reshape(X_1.shape),alpha = 0.2, cmap = ListedColormap(('tab:blue', 'tab:orange')))

markersA = plt.Line2D([0,0],[0,0],color=colors1['1'], marker=marker1['1'], linestyle='')
markersB = plt.Line2D([0,0],[0,0],color=colors1['-1'], marker=marker1['-1'], linestyle='')
proxyA = plt.Rectangle((0, 0), 1, 1, fc='tab:orange')
proxyB = plt.Rectangle((0, 0), 1, 1, fc='tab:blue')
plt.legend([markersA, markersB, proxyA, proxyB], ["Actual 1", "Actual -1", "Predicted 1", "Predicted -1"], numpoints=1, loc='upper right')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.title('Decision Boundary Plot')
plt.show()
```
